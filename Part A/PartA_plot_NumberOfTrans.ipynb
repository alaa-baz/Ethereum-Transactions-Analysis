```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('NumberOfTrans result.csv')

dataset = pd.DataFrame(data)
```

```python
x = dataset['date']
y = list(dataset['val'])

# Create bars
plt.figure(figsize=(20,12))
plt.bar(x, y)

plt.xticks(rotation= 75)
plt.xlabel('Months')
plt.ylabel('Number of Transactions')

# Plot graph
plt.show()
```
