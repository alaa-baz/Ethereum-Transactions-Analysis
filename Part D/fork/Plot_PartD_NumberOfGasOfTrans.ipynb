```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('out_avrOfGasPricOfTran_before.csv')

dataset = pd.DataFrame(data)
```

```python
x = dataset['date']
y = list(dataset['val'])

# Create bars
plt.figure(figsize=(15,10))
plt.bar(x, y)

plt.xticks(rotation= 75)
plt.xlabel('Date before Fork')
plt.ylabel('Average of gas of transactions')

# Plot graph
plt.show()
```
